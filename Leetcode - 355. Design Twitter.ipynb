```python
#Heap-PriorityQueue
class Twitter(object):
    def __init__(self):
        self.timestamp=0
        self.usertweets = {}
        self.usersfollow = {}

    def postTweet(self, userId, tweetId):
        """
        :type userId: int
        :type tweetId: int
        :rtype: None
        """
        print("postTweet")
        self.usertweets[userId] = self.usertweets.get(userId,[])
        postsFromUser = self.usertweets[userId]
        heapq.heappush(self.usertweets[userId],[self.timestamp,tweetId])
        self.timestamp-=1
        

    def getNewsFeed(self, userId):
        """
        :type userId: int
        :rtype: List[int]
        """
        print("getNewsFeed")
        self.usersfollow[userId]=self.usersfollow.get(userId,[userId])
        alltweets=[]
        for following in self.usersfollow[userId]:
            tweetsfromfollowing=self.usertweets.get(following,[])
            if len(tweetsfromfollowing)>0:
                for tweets in tweetsfromfollowing:
                    heapq.heappush(alltweets,tweets)
        res=[]
        i=0
        while alltweets and i<10:
            res.append(heapq.heappop(alltweets)[1])
            i+=1
        return res

    def follow(self, followerId, followeeId):
        """
        :type followerId: int
        :type followeeId: int
        :rtype: None
        """
        print("getFollow")
        self.usersfollow[followerId] = self.usersfollow.get(followerId,[followerId])
        following = self.usersfollow[followerId]
        if followeeId in following:
            return
        self.usersfollow[followerId].append(followeeId)
        

    def unfollow(self, followerId, followeeId):
        """
        :type followerId: int
        :type followeeId: int
        :rtype: None
        """
        print("UnFollow")
        self.usersfollow[followerId] = self.usersfollow.get(followerId,[followerId])
        following = self.usersfollow[followerId]
        if followeeId not in following or following==[followerId]:
            return
        self.usersfollow[followerId].remove(followeeId)


# Your Twitter object will be instantiated and called as such:
# obj = Twitter()
# obj.postTweet(userId,tweetId)
# param_2 = obj.getNewsFeed(userId)
# obj.follow(followerId,followeeId)
# obj.unfollow(followerId,followeeId)
```
