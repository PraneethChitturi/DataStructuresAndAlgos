```python
/**
 * Definition for a binary tree node.
 * public class TreeNode {
 *     int val;
 *     TreeNode left;
 *     TreeNode right;
 *     TreeNode() {}
 *     TreeNode(int val) { this.val = val; }
 *     TreeNode(int val, TreeNode left, TreeNode right) {
 *         this.val = val;
 *         this.left = left;
 *         this.right = right;
 *     }
 * }
 */
class Solution {
    TreeNode res=null;

    public List<TreeNode> findDeepestNodes(TreeNode root){
        List<List<TreeNode>> resultdeepnodes = new ArrayList<>();
        Queue<TreeNode> queue = new LinkedList<>();
        queue.offer(root);


        while (!queue.isEmpty()){
            int size = queue.size();
            List<TreeNode> currenlevel = new ArrayList<>();
            
            for (int i=0;i<size;i++){
                TreeNode ele = queue.poll();
                currenlevel.add(ele);
                if (ele.left!=null){
                    queue.offer(ele.left);
                }
                if (ele.right!=null){
                    queue.offer(ele.right);
                }
                
            }

            resultdeepnodes.add(currenlevel);
        }

        return resultdeepnodes.get(resultdeepnodes.size()-1);
    }

    public TreeNode subtreeWithAllDeepest(TreeNode root) {
        List<TreeNode> deepnodes = findDeepestNodes(root);
        int target = deepnodes.size();

        this.recursiveTreedeep(root,target,deepnodes);

        return this.res;
    }

    public int recursiveTreedeep(TreeNode node,int target,List<TreeNode> deepNodes){
        int left=0;
        int right=0;

        if (node.left!=null){
            left=recursiveTreedeep(node.left,target,deepNodes);
        }
        if (node.right!=null){
            right=recursiveTreedeep(node.right,target,deepNodes);
        }

        int selfcount=deepNodes.contains(node)?1:0;

        int total = left+right+selfcount;

        if (total==target && this.res==null){
            this.res=node;
        }
        return total;
    }

}
```
