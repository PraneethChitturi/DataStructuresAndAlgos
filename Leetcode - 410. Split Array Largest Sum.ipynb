```python
#DP
class Solution {
    Integer[][] memo;
    int[] prefixSum;

    public int splitArray(int[] nums, int k) {
        this.memo = new Integer[nums.length][k+1];
        this.prefixSum = new int[nums.length+1];

        for (int i=0;i<nums.length;i++){
          prefixSum[i+1] = prefixSum[i] + nums[i];
        }

        return dp(0,k,nums.length+1);
    }

    public int dp(int currentindex,int subArraycount,int lentotal){
        if (memo[currentindex][subArraycount]!=null){
            return memo[currentindex][subArraycount];
        }

        if (subArraycount==1){
            return memo[currentindex][subArraycount]=prefixSum[lentotal-1]-prefixSum[currentindex];
        }

        int minSumofCombo=Integer.MAX_VALUE;
        for (int i=currentindex; i<=(lentotal-1)-subArraycount;i++){
            int firstsplitsum=prefixSum[i+1]-prefixSum[currentindex];
            if (firstsplitsum>minSumofCombo){
                break;
            }
            int tempSumofCombo= Math.max(firstsplitsum,
                                    dp(i+1,subArraycount-1,lentotal));
            minSumofCombo = Math.min(minSumofCombo,tempSumofCombo);
        }
        memo[currentindex][subArraycount] = minSumofCombo;

        return memo[currentindex][subArraycount] = minSumofCombo;
    }

}
```

```python
#Binary Search
class Solution {

    public int splitArray(int[] nums, int k) {
        int left=0;
        int right=0;
        for (int i:nums){
            left=Math.max(left,i);
            right+=i;
        }
        
        int minimumLargestSplitSum=0;
        while (left<=right){
            int maxSumAllowed=left+(right-left)/2;

            if (countSubarraysFORmaxSumAllowed(nums,maxSumAllowed)<=k){
                right=maxSumAllowed - 1;
                minimumLargestSplitSum = maxSumAllowed;
            } else {
                left=maxSumAllowed+1;
            }

        }

        return minimumLargestSplitSum;
        
    }

    public int countSubarraysFORmaxSumAllowed(int[] nums,int permsum){
        int currentsum=0;
        int subarrayposs=0;

        for (int i:nums){
            if (currentsum+i>permsum){
                subarrayposs+=1;
                currentsum=i;
            } else {
                currentsum+=i;
            }
        }

        return subarrayposs+1;
    }

}
```
