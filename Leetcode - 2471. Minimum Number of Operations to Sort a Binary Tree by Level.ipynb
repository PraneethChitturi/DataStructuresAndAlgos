```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

from collections import deque

class Solution(object):
    def minimumOperations(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """
        if not root:
          return 0

        queue =deque([root])

        minsteps= 0
        while queue:
          currentlistlen = len(queue)
        
          listtosort = []
          
          for i in range(currentlistlen):
            currentele = queue.popleft()
            if currentele.left:
              queue.append(currentele.left)
            if currentele.right:
              queue.append(currentele.right)
            listtosort.append(currentele.val)

          minsteps = minsteps + self.minSwaps(listtosort)

        return minsteps

    def minSwaps(self, arr):
            swaps = 0
            target = sorted(arr)
            pos = {val: idx for idx, val in enumerate(arr)}

            for i in range(len(arr)):
                while arr[i] != target[i]:
                    swaps += 1
                    swap_idx = pos[target[i]]

                    # Update position map before the swap
                    pos[arr[i]] = swap_idx
                    pos[target[i]] = i

                    # Swap values
                    arr[i], arr[swap_idx] = arr[swap_idx], arr[i]

            return swaps
        
```
