```python
#Beats 58%
class MyStack(object):

    def __init__(self):
        self.stack = []
        self.len = 0
        
    def push(self, x):
        """
        :type x: int
        :rtype: None
        """
        self.stack.append(x)
        self.len += 1

    def pop(self):
        """
        :rtype: int
        """
        temp=self.stack[-1]
        self.stack=self.stack[0:-1]
        self.len -= 1
        return temp

    def top(self):
        """
        :rtype: int
        """
        return self.stack[-1]
        

    def empty(self):
        """
        :rtype: bool
        """
        return self.len==0
```
