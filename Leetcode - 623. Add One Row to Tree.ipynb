```python
#Beats 75% in TLE
class TreeNode(object):
     def __init__(self, val=0, left=None, right=None):
            self.val = val
            self.left = left
            self.right = right
            
class Solution(object):
    def addOneRow(self, root, val, depth):
        """
        :type root: TreeNode
        :type val: int
        :type depth: int
        :rtype: TreeNode
        """
        if depth==1:
            ele = TreeNode(val)
            ele.left = root
            return ele
        
        def dfs(currele,curr,val,depth,flag):
            if flag == 1:
                return currele
            
            if curr == depth-1:
                leftele = TreeNode(val)
                leftele.left = currele.left
                rightele = TreeNode(val)
                rightele.right = currele.right
                
                currele.left = leftele
                currele.right = rightele
                
                flag = 1
            
            if currele.left:
                dfs(currele.left,curr+1,val,depth,flag)
            if currele.right:
                dfs(currele.right,curr+1,val,depth,flag)
            return currele
            
        return dfs(root,1,val,depth,0)
        
```

```python
#Beats 93% in TLE
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def addOneRow(self, root, val, depth):
        """
        :type root: TreeNode
        :type val: int
        :type depth: int
        :rtype: TreeNode
        """
        if depth == 1:
            node = TreeNode(val)
            node.left = root
            return node

        def dfs(node, prev, d):
            if not node:
                return 

            if d == depth - 1:
               temp = node.left 
               node.left = TreeNode(val)
               node.left.left = temp 
               
               temp = node.right
               node.right = TreeNode(val)
               node.right.right = temp
            else:
                dfs(node.left, node, d+1)
                dfs(node.right, node, d+1)
        
        dfs(root, None, 1)
        return root
        
```
