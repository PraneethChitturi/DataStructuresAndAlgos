```python
from collections import defaultdict
from collections import deque
class Solution(object):
    def countGoodNodes(self, edges):
        """
        :type edges: List[List[int]]
        :rtype: int
        """
        def create_child_dict(edges, root):
            dictofchild = defaultdict(list)

            # Create an adjacency list
            adj_list = defaultdict(list)
            for a, b in edges:
                adj_list[a].append(b)
                adj_list[b].append(a)

            # Use a queue for BFS to properly assign children to their respective parents
            queue = deque([root])
            visited = set([root])

            while queue:
                parent = queue.popleft()
                for child in adj_list[parent]:
                    if child not in visited:
                        dictofchild[parent].append(child)
                        queue.append(child)
                        visited.add(child)

            return dictofchild
        dictofchild = create_child_dict(edges, 0)  
        #print(dictofchild)
        subtreechildren = {}
        def sizeofsubtree(nodeval):
            size=0
            for children in dictofchild[nodeval]:
                size+=sizeofsubtree(children)
            subtreechildren[nodeval]=size
            return size+1
            
        sizeofsubtree(0)
        goodnodes = 0
        for i in range(len(subtreechildren)):
            flag=True
            prevchildsize=None
            for children in dictofchild[i]:
                if prevchildsize==None:
                    prevchildsize=subtreechildren[children]
                elif prevchildsize!=subtreechildren[children]:
                    flag=False
            
            if flag==True:
                goodnodes+=1
                    
                
                
        return goodnodes
```
