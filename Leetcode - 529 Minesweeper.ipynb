```python
#Works fine, but only beats 70%
board = [["B","1","E","1","B"],["B","1","M","1","B"],["B","1","1","1","B"],["B","B","B","B","B"]]

click = [1,2]

def updateBoard(board, click):
        """
        :type board: List[List[str]]
        :type click: List[int]
        :rtype: List[List[str]]
        """
        R = len(board)
        C = len(board[0])
        if board[click[0]][click[1]] == 'M':
            board[click[0]][click[1]] = 'X'
            return board
        
        visitedmat = [[-1]*C for i in range(R)]
        queue = [[click[0],click[1]]]
        row = [0,0,-1,1,-1,1,-1,1]
        col = [-1,1,0,0,1,1,-1,-1]
        
        while queue:
            click_x,click_y = queue.pop(0)
            
            visitedmat[click_x][click_y]=1
            
            count = 0
            tempq = []
            for i in range(8):
                if click_x+col[i]<R and click_y+row[i]<C and click_x+col[i]>=0 and click_y+row[i]>=0:
                    if board[click_x+col[i]][click_y+row[i]] == 'M':
                        count = count+1
                    if (visitedmat[click_x+col[i]][click_y+row[i]]!=1) and ([click_x+col[i],click_y+row[i]] not in queue):
                        tempq.append([click_x+col[i],click_y+row[i]])
                        
            if count == 0 and board[click_x][click_y]!='M':
                board[click_x][click_y]='B'
                queue = queue+tempq
            elif board[click_x][click_y]!='M':
                board[click_x][click_y]=count
                
        return board
```

```python
#Faster Solution
class Solution(object):
    def getAdMines(self, click, board):
        mines = 0
        neighbours = []
        for i in range(click[0]-1, click[0]+2):
            for j in range(click[1]-1, click[1]+2):
                if 0 <= i <= len(board)-1 and 0 <= j <= len(board[0])-1 and (i != click[0] or j != click[1]):
                    if board[i][j] == 'MV' or board[i][j] == 'M':
                        mines += 1
                    if board[i][j] == 'M' or board[i][j] == 'E':
                        neighbours.append([i,j])
                    
        if mines == 0:
            board[click[0]][click[1]] = 'B'
            for a,b in neighbours:
                board[a][b] = board[a][b] + 'V'
        else:
            board[click[0]][click[1]] = str(mines)
        return mines, neighbours

    def updateBoard(self, board, click):
        """
        :type board: List[List[str]]
        :type click: List[int]
        :rtype: List[List[str]]
        """

        queue = [click]
        if board[click[0]][click[1]] == 'M':
            board[click[0]][click[1]] = 'X'
        elif board[click[0]][click[1]] == 'E':
            while queue:
                square = queue.pop(0)
                m, n = self.getAdMines(square, board)
                if m == 0:
                    queue = queue + n

        return board
```
