```python
#Beats 19% in TLE
class neighborSum(object):

    def __init__(self, grid):
        """
        :type grid: List[List[int]]
        """
        self.grid = grid
        self.length = len(grid)
        
    def adjacentSum(self, value):
        """
        :type value: int
        :rtype: int
        """
        for i in range(self.length):
          for j in range(self.length):
            if self.grid[i][j]==value:
              if i==0 and j>0 and j<self.length-1:
                return self.grid[i+1][j]+ self.grid[i][j-1]+ self.grid[i][j+1]
              elif i==0 and j==0:
                return self.grid[i][j+1] + self.grid[i+1][j]
              elif i==0 and j==self.length-1:
                return self.grid[i][j-1] + self.grid[i+1][j]
              elif i==self.length-1 and j>0 and j<self.length-1:
                return self.grid[i-1][j] + self.grid[i][j-1]+ self.grid[i][j+1]
              elif i==self.length-1 and j==0:
                return self.grid[i-1][j] + self.grid[i][j+1]
              elif i==self.length-1 and j==self.length-1:
                return self.grid[i-1][j] + self.grid[i][j-1]
              elif i>0 and i<self.length-1 and j>0 and j<self.length-1:
                return self.grid[i-1][j] + self.grid[i+1][j]+  self.grid[i][j-1]+ self.grid[i][j+1]
              elif j==self.length-1 and i>0 and i<self.length-1:
                  return self.grid[i][j-1] +self.grid[i-1][j]+self.grid[i+1][j]
              elif j==0 and i>0 and i<self.length-1:
                  return self.grid[i][j+1] +self.grid[i-1][j]+self.grid[i+1][j]
              return 0

    def diagonalSum(self, value):
        """
        :type value: int
        :rtype: int
        """
        for i in range(self.length):
          for j in range(self.length):
            if self.grid[i][j]==value:
              if i==0 and j>0 and j<self.length-1:
                return self.grid[i+1][j-1]+ self.grid[i+1][j+1]
              elif i==0 and j==0:
                return self.grid[i+1][j+1]
              elif i==0 and j==self.length-1:
                return self.grid[i+1][j-1]
              elif i==self.length-1 and j>0 and j<self.length-1:
                return self.grid[i-1][j-1] + self.grid[i-1][j+1]
              elif i==self.length-1 and j==0:
                return self.grid[i-1][j+1]
              elif i==self.length-1 and j==self.length-1:
                return self.grid[i-1][j-1]
              elif i>0 and i<self.length-1 and j>0 and j<self.length-1:
                  return self.grid[i-1][j-1] +self.grid[i-1][j+1]+self.grid[i+1][j+1]+self.grid[i+1][j-1]

              elif j==self.length-1 and i>0 and i<self.length-1:
                  return self.grid[i-1][j-1] +self.grid[i+1][j-1]
              elif j==0 and i>0 and i<self.length-1:
                  return self.grid[i-1][j+1] +self.grid[i+1][j+1]
              return 0


# Your neighborSum object will be instantiated and called as such:
# obj = neighborSum(grid)
# param_1 = obj.adjacentSum(value)
# param_2 = obj.diagonalSum(value)
```
